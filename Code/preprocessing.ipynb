```python
import numpy as np
import pandas as pd
```

```python
prices = pd.read_csv('Input Files/prices.csv')
```

```python
prices
```

```python
prices['date'] = pd.to_datetime(prices['date'])

# Set the 'date' column as the index for easier manipulation
prices.set_index('date', inplace=True)
```

```python
prices
```

```python
weekly_returns = prices.pct_change()*100

# Reset the index to have the 'date' column back
```

```python
weekly_returns
```

```python
weekly_volatility = weekly_returns.rolling(window=156).std()
```

```python
weekly_volatility
```

```python
import pandas as pd 
volatility = weekly_volatility

# Add a new column 'quarter' that labels the quarter for each datapoint
volatility['quarter'] = weekly_volatility.index.to_period('Q')

volatility


```

```python
# Create a new DataFrame with only the last datapoint for each quarter
vol_quarterly = volatility.resample('Q').last()
```

```python
vol_quarterly
```

```python
vol_quarterly.set_index('quarter', inplace=True)
```

```python
vol_quarterly
```

```python
vol_quarterly.to_excel('processed data/quarterly volatility.xlsx')
weekly_returns.to_excel('processed data/weekly returns.xlsx')
```

```python
prices.drop(columns=['quarter'], inplace=True)

price_4_weeks_ago = prices.shift(4)
price_52_weeks_ago = prices.shift(52)

# Calculate momentum as the ratio of price 4 weeks ago to price 52 weeks ago
momentum = price_4_weeks_ago / price_52_weeks_ago

momentum['quarter'] = momentum.index.to_period('Q')
mom_quarterly = momentum.resample('Q').last()
mom_quarterly.set_index('quarter', inplace=True)
mom_quarterly
```

```python
mom_quarterly.to_excel('processed data/quarterly momentum.xlsx')
```

```python
monthly_div = pd.read_excel('Input Files/dividend_yield.xlsx')
```

```python
dividends = monthly_div
```

```python


dividends['Dates'] = pd.to_datetime(dividends['Dates'])
dividends.set_index('Dates', inplace=True)

dividends['quarter'] = dividends.index.to_period('Q')

div_quarterly = dividends.resample('Q').last()
div_quarterly.set_index('quarter', inplace=True)
div_quarterly
```

```python
outstanding = pd.read_csv('Input Files/outstanding_shares.csv')
outstanding['Dates'] = pd.to_datetime(outstanding['Dates'])
outstanding.set_index('Dates', inplace=True)

rolling_24_month_avg = outstanding.rolling(window=24, min_periods=24).mean()

# Calculate the buyback yield
buyback_yield = ((outstanding - rolling_24_month_avg) / rolling_24_month_avg)*-1
```

```python
buyback_yield
```

```python
buyback_yield['quarter'] = buyback_yield.index.to_period('Q')

bb_quarterly = buyback_yield.resample('Q').last()
bb_quarterly.set_index('quarter', inplace=True)
bb_quarterly
```

```python
npy_quarterly = bb_quarterly + div_quarterly
npy_quarterly
```

```python
bb_quarterly.to_excel('processed data/quarterly buyback yield.xlsx')
npy_quarterly.to_excel('processed data/quarterly net payout yield.xlsx')
```
